```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Load the data from a CSV file
data = pd.read_csv("survey_data.csv")

# Perform sentiment analysis and store results in a new list
positive_count = 0
neutral_count = 0
negative_count = 0

for index, row in data.iterrows():
    text = str(row['Have you ever experienced a data breach or privacy violation when shopping online? If so, how did you feel about it and how did the company respond?'])
    sentiment = TextBlob(text).sentiment.polarity
    
    if sentiment > 0:
        positive_count += 1
    elif sentiment < 0:
        negative_count += 1
    else:
        neutral_count += 1

# Create a pie chart
labels = ['Positive', 'Neutral', 'Negative']
sizes = [positive_count, neutral_count, negative_count]
colors = ['green', 'yellow', 'red']
explode = (0.1, 0, 0)  # explode the 1st slice (Positive)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a title
plt.title('Sentiment Analysis in Interview Responses')

# Show the pie chart
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("survey_data.csv")

# Select the column containing the text responses
text_responses = data['How does it make you feel to know that your data is being used without your knowledge or consent?']

# Text processing
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm = vectorizer.fit_transform(text_responses)

# Topic modeling using Latent Dirichlet Allocation
num_topics = 3
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(dtm)

# Display the top words for each topic
def display_topics(model, feature_names, num_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx + 1))
        print(" ".join([feature_names[i] for i in topic.argsort()[:-num_top_words - 1:-1]]))
        print()

num_top_words = 5
feature_names = vectorizer.get_feature_names_out()
display_topics(lda, feature_names, num_top_words)

# Assign themes based on top words
themes = {
    'Theme 1': ['privacy', 'concerned', 'data', 'important', 'security'],
    'Theme 2': ['transparency', 'business', 'model', 'responsibility', 'trust'],
    'Theme 3': ['authentic', 'products', 'responsibility', 'customer', 'safety']
}

# Manual coding of responses
coded_data = []
for idx, row in data.iterrows():
    response = row['How does it make you feel to know that your data is being used without your knowledge or consent?']
    theme_scores = {theme: sum(response.lower().count(word) for word in words) for theme, words in themes.items()}
    coded_data.append({
        'Response': response,
        'Themes': theme_scores
    })

# Display coded data
for entry in coded_data:
    print("Response:", entry['Response'])
    for theme, score in entry['Themes'].items():
        print(f"{theme}: {score}")
    print("-" * 50)

# Visualization of theme distribution
theme_labels = list(themes.keys())
theme_scores = pd.DataFrame([entry['Themes'] for entry in coded_data])
theme_scores.plot(kind='bar', stacked=True)
plt.xticks(range(len(coded_data)), range(1, len(coded_data) + 1), rotation=0)
plt.xlabel('Response')
plt.ylabel('Theme Score')
plt.title('Theme Distribution in Responses')
plt.legend(theme_labels)
plt.show()
```

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Load the survey data from a CSV file
data = pd.read_csv("survey_data.csv")

# Define the column containing text data for analysis
text_column = 'In your opinion, should there be more regulation in place to ensure that online stores and dropshippers are handling customer data legally? Why or why not?'

# Perform sentiment analysis and store results in a new column
data['Sentiment'] = data[text_column].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Define a function to categorize sentiment
def categorize_sentiment(sentiment_score):
    if sentiment_score > 0:
        return 'Positive'
    elif sentiment_score < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment categorization to create a new column
data['Sentiment Category'] = data['Sentiment'].apply(categorize_sentiment)

# Calculate sentiment distribution
sentiment_distribution = data['Sentiment Category'].value_counts()

# Create a bar plot to visualize sentiment distribution
plt.bar(sentiment_distribution.index, sentiment_distribution.values, color=['green', 'red', 'yellow'])
plt.title('Sentiment Analysis in Survey Responses')
plt.xlabel('Sentiment Category')
plt.ylabel('Number of Responses')
plt.show()

# Calculate average sentiment score
average_sentiment = data['Sentiment'].mean()
print(f'Average Sentiment Score: {average_sentiment:.2f}')

```
